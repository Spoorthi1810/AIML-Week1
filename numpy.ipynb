```python
#1 
a = np.array([1,2,3])
b = np.array([4,5,6])

print("Array a:", a)
print("Array b:", b)

# Addition
print("Addition:", a + b)

# Subtraction
print("Subtraction:", a - b)

# Multiplication
print("Multiplication:", a * b)

# Division
print("Division:", a / b)
```
