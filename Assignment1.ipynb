```python
# 14 Introduce missing values in Marks column.

df2.loc[0, "Marks"] = None

print("Data After Introducing Missing Value:\n")
print(df2.head())
```

```python
# 15 Identify and replace missing values with average marks.

print("Missing Values Before:\n", df2.isnull().sum())

df2["Marks"].fillna(df2["Marks"].mean(), inplace=True)

print("\nMissing Values After:\n", df2.isnull().sum())
```

```python
# 16 Check and remove duplicate rows.

print("Duplicate Rows Before:", df2.duplicated().sum())

df2.drop_duplicates(inplace=True)

print("Duplicate Rows After:", df2.duplicated().sum())
```

```python
# 17 Create Result column (Pass if Marks >= 40 else Fail).

df2["Result"] = df2["Marks"].apply(lambda x: "Pass" if x >= 40 else "Fail")

print(df2)
```

```python
# 18 Rename Marks column to Final_Marks.

df2.rename(columns={"Marks": "Final_Marks"}, inplace=True)

print(df2.head())
```

```python

# 19 Drop Subject column.

df2.drop("Subject", axis=1, inplace=True)

print(df2.head())
```

```python
# 20 Export final DataFrame to CSV and read it back.

df2.to_csv("final_output.csv", index=False)

new_df = pd.read_csv("final_output.csv")

print("Re-imported Data:\n")
print(new_df.head())
```

```python
# Bonus 1 Find students having the same highest marks.

# Find highest marks
highest_marks = df2["Final_Marks"].max() if "Final_Marks" in df2.columns else df2["Marks"].max()

# Filter students who have highest marks
top_students = df2[df2["Final_Marks"] == highest_marks] if "Final_Marks" in df2.columns else df2[df2["Marks"] == highest_marks]

print("Highest Marks:", highest_marks)
print("Students with Highest Marks:\n")
print(top_students)
```

```python
# Bonus 2 Create a Grade column based on marks.

# Define grading function
def grade(m):
    if m >= 85:
        return "A"
    elif m >= 70:
        return "B"
    elif m >= 50:
        return "C"
    else:
        return "D"

# Apply grading
if "Final_Marks" in df2.columns:
    df2["Grade"] = df2["Final_Marks"].apply(grade)
else:
    df2["Grade"] = df2["Marks"].apply(grade)

print("Data with Grade Column:\n")
print(df2)
```

```python
# Bonus 3 Calculate and display pass percentage.

# Create Result column if not already created
if "Result" not in df2.columns:
    if "Final_Marks" in df2.columns:
        df2["Result"] = df2["Final_Marks"].apply(lambda x: "Pass" if x >= 40 else "Fail")
    else:
        df2["Result"] = df2["Marks"].apply(lambda x: "Pass" if x >= 40 else "Fail")

# Calculate pass percentage
pass_count = df2[df2["Result"] == "Pass"].shape[0]
total_students = df2.shape[0]

pass_percentage = (pass_count / total_students) * 100

print("Pass Percentage:", pass_percentage, "%")
```
