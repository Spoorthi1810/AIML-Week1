```python
# 1. Create a NumPy array of numbers from 1 to 20

arr = np.arange(1, 21)
print(arr)
```

```python
# 2. Reshape array into 4x5 matrix

matrix = arr.reshape(4, 5)
print("4x5 Matrix:\n", matrix)
```

```python
# 3. Find Mean, Median and Standard Deviation

print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
```

```python
# 4. Extract all even numbers

even_numbers = arr[arr % 2 == 0]
print("Even Numbers:", even_numbers)
```

```python
# Find the student with highest average marks.
import pandas as pd

df = pd.read_csv("student_marks_demo.csv")

# Convert Marks properly
df['Marks'] = pd.to_numeric(df['Marks'], errors='coerce')

# Remove rows where Marks became NaN
df = df.dropna(subset=['Marks'])

# Now calculate average per student
student_avg = df.groupby('Name', as_index=False)['Marks'].mean()

print("Average Marks Table:")
print(student_avg)

# Now safely get highest student
top_row = student_avg.sort_values(by='Marks', ascending=False).iloc[0]

print("\nTop Student:")
print(top_row)
```

```python

# 6. Students who scored more than 75 in Math


math_students = df[(df['Subject'] == 'Math') & (df['Marks'] > 75)]

print(math_students)
```

```python
# 7 Add a new column 'Result' (Pass/Fail) based on average >= 40 
avg_marks = df.groupby('Name')['Marks'].mean().reset_index()
avg_marks.rename(columns={'Marks':'Average'}, inplace=True)

# Add Result column
avg_marks['Result'] = avg_marks['Average'].apply(lambda x: 'Pass' if x >= 40 else 'Fail')

print(avg_marks)
```

```python
#Section 3: Matplotlib (Visualization Tasks)  

# =====================================================
# Calculate average marks per student
# =====================================================

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student_marks_demo.csv")

df['Marks'] = pd.to_numeric(df['Marks'], errors='coerce')
df = df.dropna(subset=['Marks'])

avg_marks = df.groupby('Name')['Marks'].mean().reset_index()
# =====================================================
# Bar Chart: Student Names vs Average Marks
# =====================================================

plt.figure()

plt.bar(avg_marks['Name'], avg_marks['Marks'])

plt.title("Student Names vs Average Marks")
plt.xlabel("Student Name")
plt.ylabel("Average Marks")

plt.xticks(rotation=45)

plt.show()

```

```python
# =====================================================
# Prepare data for graphs
# =====================================================

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student_marks_demo.csv")

# clean marks
df['Marks'] = pd.to_numeric(df['Marks'], errors='coerce')
df = df.dropna(subset=['Marks'])

# create average table
avg_marks = df.groupby('Name')['Marks'].mean().reset_index()
avg_marks.rename(columns={'Marks':'Average'}, inplace=True)

print(avg_marks)   # check column exists
```

```python
# =====================================================
#2. Pie chart: Pass vs Fail
# =====================================================

avg_marks['Result'] = avg_marks['Average'].apply(lambda x: 'Pass' if x >= 40 else 'Fail')

result_count = avg_marks['Result'].value_counts()

plt.figure()
plt.pie(result_count, labels=result_count.index, autopct='%1.1f%%')
plt.title("Pass vs Fail Distribution")
plt.show()
```

```python

# 3.Line graph subject comparison
# =====================================================
# Perfect comparison graph (only students with all 3 subjects)
# =====================================================

plt.figure(figsize=(8,5))

for subject in df['Subject'].unique():
    subset = df[df['Subject'] == subject]
    plt.plot(subset['Name'], subset['Marks'], marker='o', label=subject)

plt.title("Marks Comparison (Math, Science, English)")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()


```
